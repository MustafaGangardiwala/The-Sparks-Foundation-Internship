### Importing the Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
# Comment this if the data visualisations doesn't work on your side
%matplotlib inline

plt.style.use('bmh')
```

### Reading the dataset

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### Checking for null values

```python
df.isnull().sum()
```

### Removing duplicate entries

```python
duplicate = df.duplicated()
print(duplicate.sum())
df[duplicate]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
col=['Postal Code']
df = df.drop(columns=col,axis=1)
```

```python
correlation=df.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)
```

```python
sns.pairplot(df, hue='Ship Mode')
```

```python
df.hist(bins=50 ,figsize=(20,15))
plt.show();
```

```python
plt.figure(figsize = (30, 12))
sns.countplot(x = 'Sub-Category', hue = 'Region', data = df, palette = 'plasma')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Sub-Categories', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

```python

plt.figure(figsize = (30, 12))
sns.countplot(x = 'Segment', hue = 'Region', data = df, palette = 'plasma')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Sub-Categories', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

```python
plt.figure(figsize=(15,10))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
sns.set(style="white")
plt.figure(2, figsize=(15,8))

sns.barplot(x='Sub-Category',y='Profit', data=df, palette='Spectral')

plt.suptitle('Pie Consumption Patterns in the United States', fontsize=16)
```

## EDA 

```python
Sales_and_Profits = df.groupby("Segment").sum().iloc[:,[1,-1]].sum()
round(Sales_and_Profits,2) # Rounding Numericala entries to 2 decimal places
```

```python
Top_10_Sales = df.groupby("State").Sales.sum().nlargest(n =10)
Top_10_Profits = df.groupby("State").Profit.sum().nlargest(n =10)
```

### Top 10 States by Sales

```python
Top_10_Sales.index
```

```python
plt.style.use('classic')
Top_10_Sales.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Sales",fontsize =13)
plt.title("Top 10 States by Sales",fontsize =16)
plt.show()
```

### Top 10 States by Profits

```python
Top_10_Profits.index
```

```python
plt.style.use('classic')
Top_10_Profits.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Profits",fontsize =13)
plt.title("Top 10 States by Profits",fontsize =16)
plt.show()
```

## Categories classified according to the region.

```python
plt.style.use('classic')
plt.figure(figsize = (30, 12))
sns.countplot(x = 'Category', hue = 'Region', data = df, palette = 'plasma')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Categories', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

```python
plt.figure(figsize = (15,6))
sns.set(font_scale=1, palette= "plasma")
sns.barplot(data = df , x = "Region",y = "Profit" ,hue = "Category")
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.set(font_scale=1, palette= "plasma")
sns.barplot(data = df , x = "State",y = "Profit" ,hue = "Category")
plt.title("Investigation of Central Region Furniture Category: Profit Analysis(by Sub Category)", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = df , x = "State",y = "Discount" ,hue = "Category")
plt.title("Discounts provided by each state", fontsize = 20)
plt.show()
```
