## EDA 

```python
correlation=df.corr()
sns.set(style='white')
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)
```

```python
df.hist(bins=50 ,figsize=(15,15))
plt.show();
```

```python
sns.countplot(x=df['Category'])
```

```python
sns.pairplot(df, hue='Category')
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=df['State'],palette='plasma_r', order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

#### California, New York and Texas have high number of sales.
#### North Dakota, West Virginia and Wyoming have low number of sales.

```python
plt.figure(figsize = (30, 12))
sns.countplot(x = 'Sub-Category', hue = 'Region', data = df, palette = 'plasma')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Sub-Categories', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

```python
plt.figure(figsize=(15,10))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

#### The most bought Sub-Categories are Binders and Paper throughout all regions.
#### While Copiers and Machines are less bought.

```python
plt.figure(figsize=(15,10))
df['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize = (30, 12))
sns.countplot(x = 'Segment', hue = 'Region', data = df, palette = 'plasma')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Segment', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

```python
Sales_and_Profits = df.groupby("Segment").sum().iloc[:,[1,-1]].sum()
round(Sales_and_Profits,2) 
```

```python
Top_10_Sales = df.groupby("State").Sales.sum().nlargest(n =10)
Top_10_Profits = df.groupby("State").Profit.sum().nlargest(n =10)
```

### Top 10 States by Sales

```python
Top_10_Sales.index
```

```python
Top_10_Sales.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Sales",fontsize =13)
plt.title("Top 10 States by Sales",fontsize =16)
plt.show()
```

### Top 10 States by Profits

```python
Top_10_Profits.index
```

```python
Top_10_Profits.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Profits",fontsize =13)
plt.title("Top 10 States by Profits",fontsize =16)
plt.show()
```

## Categories classified according to the region.

```python
plt.figure(figsize = (30, 12))
sns.countplot(x = 'Category', hue = 'Region', data = df, palette = 'plasma')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Categories', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

```python
sns.set(style='white')
plt.figure(figsize = (15,6))
sns.set(font_scale=1, palette= "plasma")
sns.barplot(data = df , x = "Region",y = "Profit" ,hue = "Category")
plt.show()
```

```python
sns.set(style='white')
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(figsize=(10,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

#### Profit and sales are maximum in consumer segment and minimum in Home Office segment

```python
sns.set(style='white')
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(figsize=(10,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

#### West Region is most profitable region

```python
sns.set(style='white')
ProfitSales = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ProfitSales[:].plot.bar(figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

#### Stated with high profit are California and New York
#### States with high loss are Texas, Illinos, Pennsylvania and Ohio

```python
sns.set(style='white')
ProfitSales = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ProfitSales[:].plot.bar(figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

#### 1) Sales of phones and chairs are close but phones are more profitable compared to Chairs.
#### 2) Although, having strong sales tables are still making loss
#### 3) Copiers having half the number of sales than phones but are almost double profitable
